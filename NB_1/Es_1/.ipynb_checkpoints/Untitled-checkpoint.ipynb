```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# Initialize empty arrays for each column
column1 = []
column2 = []
column3 = []

# Read the file
with open('data.txt', 'r') as file:
    for line in file:
        # Split each line into columns
        data = line.split()
        # Append each column to its corresponding array
        column1.append(int(data[0]))
        column2.append(float(data[1]) - 0.5)
        column3.append(float(data[2]))


plt.errorbar(column1, column2 , column3)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```
