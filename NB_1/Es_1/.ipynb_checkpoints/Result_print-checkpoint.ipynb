```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# Initialize empty arrays for each column
N_trows = []
cumul_ave = []
error_cumul_ave = []

# Read the file
with open('data.txt', 'r') as file:
    for line in file:
        # Split each line into columns
        data = line.split()
        # Append each column to its corresponding array
        N_trows.append(int(data[0]))
        cumul_ave.append(float(data[1]) - 0.5)
        error_cumul_ave.append(float(data[2]))


plt.errorbar(N_trows, cumul_ave , error_cumul_ave)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def read_file(filename):
    data = []
    with open(filename, 'r') as file:
        for line in file:
            # Assuming one column in the file
            value = float(line.strip())  # Assuming each line contains a single value
            data.append(value)
    return data

def plot_histogram(data):
    plt.hist(data, bins=20)  # You can adjust the number of bins as per your data
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title('Histogram')
    plt.show()

if __name__ == "__main__":
    filename = 'isto_chi_squared.txt'  # Change this to your file name
    data = read_file(filename)
    plot_histogram(data)
```
