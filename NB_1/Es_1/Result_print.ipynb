```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# Initialize empty arrays for each column
N_trows = []
cumul_ave = []
error_cumul_ave = []

# Read the file
with open('data.txt', 'r') as file:
    for line in file:
        # Split each line into columns
        data = line.split()
        # Append each column to its corresponding array
        N_trows.append(int(data[0]))
        cumul_ave.append(float(data[1]) - 0.5)
        error_cumul_ave.append(float(data[2]))


plt.errorbar(N_trows, cumul_ave , error_cumul_ave)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```
