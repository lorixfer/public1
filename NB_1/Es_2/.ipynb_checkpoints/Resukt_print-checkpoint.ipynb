```python
import matplotlib.pyplot as plt

def read_file(filename):
    data = []
    with open(filename, 'r') as file:
        for line in file:
            # Assuming one column in the file
            value = float(line.strip())  # Assuming each line contains a single value
            data.append(value)
    return data

def plot_histogram(data):
    plt.hist(data, bins=1000)  # You can adjust the number of bins as per your data
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title('Histogram')
    #plt.xlim(-250, 250)
    plt.show()

if __name__ == "__main__":
    filename = 'Lin1.txt'  # Change this to your file name
    data = read_file(filename)
    plot_histogram(data)
```

```python
def plot_histogram(data):
    plt.hist( data , bins=1000)  # You can adjust the number of bins as per your data
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title('Histogram')
    plt.xlim(-30, 30)
    plt.show()

def read_file(filename):
    data = []
    with open(filename, 'r') as file:
        for line in file:
            # Assuming one column in the file
            value = float(line.strip())  # Assuming each line contains a single value
            if value > -100 and value < 100: 
                data.append(value)
    return data


if __name__ == "__main__":
    filename = 'C_L1.txt'  # Change this to your file name
    data = read_file(filename)
    plot_histogram(data)
```
